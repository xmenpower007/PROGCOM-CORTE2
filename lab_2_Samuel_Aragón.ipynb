###Introducción:
>En estas dos prácticas se trabajará sobre los segmentos de recta en el espacio, es decir,
sobre los vectores. Este es un tema con bastante relevancia, esto porque no se
comportan como números escalares; estos tienen sus operaciones propias. Los vectores
tienen funciones muy útiles, las cuales utilizamos diariamente: son funciones que nos
indican la dirección, la intensidad, la fuerza, la velocidad o el desplazamiento. En estos
laboratorios será necesario hallar longitudes de un vector en el plano cartesiano para el
eje x y el eje y. En un principio debemos obtener el ángulo entre el vector y el eje x
positivó, a ese valor le llamaremos Θ, para después reemplazarlo en estas fórmulas:

>Primeramente, se observa la fórmula que nos permite hallar la longitud del eje x.
* X = Cos^-1 (Θ) (1)

>Θ = Ángulo que forma el vector con el eje x positivo.
Posterior, se toma una fórmula distinta que nos permite hallar la longitud del eje y.

* Y = Sen^-1 (Θ). (2)

>Θ = Ángulo que forma el vector con el eje x positivo.


>Se hace importancia en que se deben hallar con el ángulo que se forma siempre con el
eje x positiva, es decir, siempre tomándolo como punto de inicio, esto para poder usar
correctamente las fórmulas (1) y (2), anotadas a priori.

>Finalmente, se buscará en este laboratorio determinar ciertas longitudes sin ángulo
alguno, este proceso se hará teniendo en cuenta que la sumatoria de fuerzas de un
producto en reposo brinda el número 0. Por lo tanto, buscaremos que número
determinara las longitudes, de x y y, del ángulo restante para que este permanezca en
el origen.

###Discusión:
>Debido a los distintos resultados, podemos evidenciar como la magnitud de cada vector
puede afectar al vector resultante, esto debido a que la variación de solo una magnitud,
va a hacer que la suma, se genere o un vector más largo o más corto donde también se
tiene en cuenta una aceleración adicional, que es la de la gravedad, ya que si esta se
tomara como si no existirá, el vector que se genera sería totalmente distinto, de igual
forma pasaría si se utiliza la gravedad de otro país, no se generaría un cambio a simple
vista, pero este ya sería un vector diferente.

>Esto también ocurre con los ángulos de los vectores, ya que si este varia, la dirección
del vector resultante también puede variar, esto se puede demostrar de una mejor
manera mediante el método gráfico, debido a que los vectores van a quedar con una
ubicación diferente a la original.
